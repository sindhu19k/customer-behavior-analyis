```python
import pandas as pd

df = pd.read_csv("customer data.csv")
df.head()
```

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
```

```python
df.info()
```

```python
df['Month'] = df['Purchase Date'].dt.month
df['Year'] = df['Purchase Date'].dt.year
```

```python
customer_spend = df.groupby('Customer ID')['Total Purchase Amount'].sum().sort_values(ascending=False)
customer_spend.head()
```

```python
customer_freq = df['Customer ID'].value_counts()
customer_freq.head()
```

```python
df['Product Category'].value_counts()
```

```python
df['Churn'].value_counts()
```

```python
monthly_sales = df.groupby('Month')['Total Purchase Amount'].sum()

import matplotlib.pyplot as plt
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.show()
```

```python
rfm = df.groupby('Customer ID').agg({
    'Purchase Date': lambda x: (df['Purchase Date'].max() - x.max()).days,
    'Customer ID': 'count',
    'Total Purchase Amount': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head()
```

```python
rfm['Segment'] = 'Low Value'

rfm.loc[rfm['Monetary'] > 5000, 'Segment'] = 'High Value'
rfm.loc[rfm['Frequency'] > 10, 'Segment'] = 'Loyal'
```

```python
rfm['Segment'].value_counts().plot(kind='bar')
plt.title("Customer Segments")
plt.show()
```

```python
rfm.groupby('Segment').mean()
```

```python
rfm['Segment'].value_counts()
```

```python
import matplotlib.pyplot as plt

rfm['Segment'].value_counts().plot(kind='bar')
plt.title("Customer Segments Distribution")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")
plt.show()
```

```python
rfm.groupby('Segment')['Monetary'].mean().plot(kind='bar')
plt.title("Average Spending by Segment")
plt.show()
```
